## 11. What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

```python
sum_of_animal_fat = df.groupby('Item').agg({'Y2014': ['sum'],'Y2017':['sum']})
print(sum_of_animal_fat)
```


## 12.What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
grouped_mean  = df['Y2015'].mean()
grouped_mean_final = np.round(grouped_mean,decimals= 3)
print(grouped_mean_final)

grouped_std  =df['Y2015'].std()
grouped_std_final = np.round(grouped_std,decimals= 3)
print(grouped_std_final)
```

## 13.What is the total number and percentage of missing data in 2016 to 2 decimal places

```python

total_number_missing_data = df['Y2016'].isnull().sum()   #Total number is always calculated using sum
print(total_number_missing_data)

total_percentage = np.round(((total_number_missing_data/len(df['Y2016']))*100),decimals=2)
print(total_percentage)

```

##14.Which year had the highest correlation with ‘Element Code’? 

```python
df.corr(method='kendall')['Element Code']

# So the answer is Y2017 
```

##15. What year has the highest sum of Import Quantity? Hint- Perform a groupby operation on ‘Element’ and use the resulting Dataframe to answer the question

```python
df.groupby('Element').agg({'Y2014': ['sum'],'Y2015':['sum'],'Y2016': ['sum'],'Y2017':['sum'],'Y2018': ['sum']})

# So the answer is 2017 has highest sum of Import Quantity
```

## 16.What is the total number of the sum of Production in 2014?

```python
df.groupby('Element').agg({'Y2014':['sum']})

#So the answer is 1,931,287.750
```

##17.Which of these elements had the highest sum in 2018?

```python
df.groupby('Element').agg({'Y2018':'max'}).sort_values(by='Element')

#So the answer is Domestic Supply quantity
```

##18.Which of these elements had the 3rd lowest sum in 2018?

```python
df.groupby('Element').agg({'Y2018': ['sum']}).sort_values

# So the answer is Protein supply quantity (g/capita/day)    11,833.560
```

##19.What is the total Import Quantity in Algeria in 2018?

```python
#extract data from dataframe
#copy the df
df_copy = df.copy()
df_copy.head(10)
df_copy_new = df_copy.drop(['Area Code','Item Code','Item','Element Code','Unit','Y2014','Y2015','Y2016','Y2017'],axis=1)
df_copy_new.head()

df_copy_new[df_copy_new['Element'] == 'Import Quantity'].groupby('Area')['Y2018'].sum()

# So the answer is Algeria    36,238.290
```
